```python
import pandas as pd
from transformers import pipeline
import nltk
```

```python
# Load the summarization model
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```python
text = """
Product management is an organizational function within a company dealing with the planning, forecasting, production, and marketing of a product at all stages of the product lifecycle.
It provides critical guidance to an organization to ensure that a product achieves its goals.
Product managers are responsible for delivering a product that aligns with customer needs and company objectives.
"""

# Generate summary
summary = summarizer(text, max_length=50, min_length=25, do_sample=False)
print("Original Text:\n", text)
print("\nSummarized Text:\n", summary[0]['summary_text'])
```

```python
# Load the sentiment analysis model
sentiment_analyzer = pipeline("sentiment-analysis")

# Analyze sentiment of the original text
sentiment = sentiment_analyzer(text)[0]
print("Sentiment:", sentiment['label'], "| Score:", sentiment['score'])
```

```python
def summarize_and_analyze(text):
    # Generate summary
    summary = summarizer(text, max_length=50, min_length=25, do_sample=False)[0]['summary_text']

    # Analyze sentiment
    sentiment = sentiment_analyzer(text)[0]

    # Print results
    print("Original Text:\n", text)
    print("\nSummarized Text:\n", summary)
    print("Sentiment:", sentiment['label'], "| Score:", sentiment['score'])

    return {"Original Text": text, "Summary": summary, "Sentiment": sentiment['label'], "Score": sentiment['score']}

# Test the function
summarize_and_analyze(text)
```

```python
# Sample feedback list
feedback_list = [
    "CHA Hong Kong Eating House: The authentic Cantonese dishes at CHA are a delight, with the Konjee Crispy Lamb being a standout. The vibrant ambiance and diverse cocktail menu make it a must-visit.",
    "Narrative: Narrative offers a fusion of traditional Indian flavors with modern techniques, featuring dishes like Salmon Kokam Ceviche. The rooftop space provides a stunning view of Pune's skyline.",
    "Gather: Gather's Mediterranean-inspired menu, including the Farmer’s Market Pizza, is perfect for sharing. The cozy atmosphere and inventive cocktails enhance the dining experience",
    "Costa Coffee: This is a new place where you will get variety of new coffess, but most of them are bland, very watery, no taste at all."

]
# Apply summarization and sentiment analysis to each feedback
results = [summarize_and_analyze(feedback) for feedback in feedback_list]

# Convert results to DataFrame for better visualization
df_results = pd.DataFrame(results)
df_results
```
